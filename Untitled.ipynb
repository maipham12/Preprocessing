## 1. Import Libraries  
**Purpose & Method:**  
- Import essential modules for file handling, data manipulation, visualization, machine learning utilities, resampling, and image inspection.  
- Use `pathlib.Path` for robust path handling and `joblib` for serializing fitted transformers.  

**Expected Outcome:**

```python
# Standard library imports
import os
from pathlib import Path
import joblib  # For serializing objects

# Data handling
import pandas as pd

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Machine learning utilities
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_extraction import FeatureHasher
from sklearn.model_selection import train_test_split

# Resampling techniques
from imblearn.over_sampling import SMOTE, RandomOverSampler
from imblearn.pipeline import Pipeline as ImbPipeline

# Image inspection
from PIL import Image

# Warnings management
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

## 2. Data Collection & Ingestion  
**Purpose & Method:**  
- Load `meta_train.csv` into a DataFrame.  
- Construct full image paths via `base_path / label / image_id`.  
- Verify that **0 image files are missing** and **0 duplicate IDs remain**, ensuring data integrity before analysis.  


```python
# Reason: Early validation of file paths and duplicate entries prevents downstream errors and ensures data integrity.
meta = pd.read_csv('meta_train.csv')
base_path = Path('train_images')
meta['image_path'] = meta.apply(lambda r: base_path / r['label'] / r['image_id'], axis=1)

# Explicitly check and raise exception for missing image files
missing = meta.loc[~meta['image_path'].apply(lambda p: p.exists()), ['image_id', 'image_path']]
if not missing.empty:
    raise FileNotFoundError(f"Missing image files detected:\n{missing.to_string(index=False)}")

# Remove duplicate image IDs to avoid bias
dupe_count = meta['image_id'].duplicated().sum()
if dupe_count > 0:
    meta = meta.drop_duplicates(subset='image_id')
    print(f"Dropped {dupe_count} duplicate entries.")
```

- **a. Class Distribution**: calculate counts and imbalance ratio. *(Class Distribution Table)*  
- **b. Age Distribution**: summarize mean, standard deviation, IQR, and range. *(Age Distribution Table)*  
- **c. Variety Distribution**: assess frequency of each rice variety. *(Variety Distribution Table)*  
- **d. Interaction Plot**: examine `age` × `variety` × `label` to identify vulnerability windows. *(Age×Variety×Label Interaction Table)*

**Key Insights:**  
a. `bacterial_leaf_blight` onset concentrated at 50–65 days; `brown_spot` and `blast` span 45–75 days.

c. Specific varieties show disease only within narrow age ranges.  
d. Late-season diseases (`dead_heart`, `bacterial_panicle_blight`) have higher mean age with low variance.  
e. Severe imbalance in both class and variety distributions.

```python
'''Cell 3: Exploratory Data Analysis (EDA)'''
# Reason: Thorough EDA uncovers underlying patterns, informs feature engineering, and identifies potential pitfalls.
print("Dataset summary:\n", meta.describe(include='all'), "\n")

# Class Distribution and Imbalance Ratio
plt.figure(figsize=(8,6))
sns.countplot(y='label', data=meta, order=meta['label'].value_counts().index)
plt.title('Class Distribution'); plt.tight_layout(); plt.show()
counts = meta['label'].value_counts()
print("Class counts:\n", counts, "\n")
imbalance_ratio = counts.max()/counts.min()
print(f"Imbalance Ratio (max/min): {imbalance_ratio:.2f}\n")

# Age Distribution and Pairwise Relationship
plt.figure(figsize=(8,5))
sns.histplot(meta['age'], bins=30, kde=True)
plt.title('Age Distribution'); plt.tight_layout(); plt.show()

plt.figure(figsize=(10,6))
sns.scatterplot(x='age', y='variety', hue='label', data=meta, alpha=0.7)
plt.title('Age vs. Variety by Label'); plt.tight_layout(); plt.show()

# Sample Images per Class
# Reason: Display a grid of one sample per class to inspect intra-class variation.
classes = meta['label'].unique()
n_cols = 5
from math import ceil
n_rows = ceil(len(classes) / n_cols)
fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols*3, n_rows*3))
for idx, cls in enumerate(classes):
    r, c = divmod(idx, n_cols)
    ax = axes[r, c] if n_rows>1 else axes[c]
    img_path = meta.loc[meta['label'] == cls, 'image_path'].iloc[0]
    img = Image.open(img_path)
    ax.imshow(img)
    ax.set_title(cls)
    ax.axis('off')
# Hide any unused subplots
for idx in range(len(classes), n_rows*n_cols):
    r, c = divmod(idx, n_cols)
    ax = axes[r, c] if n_rows>1 else axes[c]
    ax.axis('off')
plt.suptitle('Sample Images per Class')
plt.tight_layout()
plt.show()

# Variety Distribution
plt.figure(figsize=(10,6))
sns.countplot(y='variety', data=meta, order=meta['variety'].value_counts().index)
plt.title('Variety Distribution'); plt.tight_layout(); plt.show()
print("Top Varieties:\n", meta['variety'].value_counts().head(), "\n")

# Age Range by Variety
var_stats = meta.groupby('variety')['age'].agg(['min','max','count']).sort_values('count', ascending=False)
print("Age ranges by variety (top 5):\n", var_stats.head(), "\n")
```

## 4. Data Cleaning & Outlier Removal  
**Purpose & Method:**  
- Apply the IQR method to `age` (retain 45–82 days), removing **0 samples**.  
- Check image file sizes (mean ~78 746 bytes); **no zero-byte files** detected.  

**Results:**  
- Cleaned DataFrame with consistent `age` range and valid file metadata.


```python
# Reason: IQR-based removal is robust for non-normal distributions in biological data.
orig_min, orig_max = meta['age'].min(), meta['age'].max()
print(f"Age range before cleaning: {orig_min} - {orig_max}")
Q1, Q3 = meta['age'].quantile([0.25, 0.75])
IQR = Q3 - Q1
low, high = Q1 - 1.5*IQR, Q3 + 1.5*IQR
outliers = meta[(meta['age']<low) | (meta['age']>high)]
out_pct = len(outliers)/len(meta)*100
print(f"Identified {len(outliers)} outliers ({out_pct:.2f}%).")
meta_clean = meta[(meta['age']>=low) & (meta['age']<=high)].copy()
clean_min, clean_max = meta_clean['age'].min(), meta_clean['age'].max()
print(f"Age range after cleaning: {clean_min} - {clean_max}\n")

# File-size anomaly check
meta_clean['file_size'] = meta_clean['image_path'].apply(lambda p: p.stat().st_size)
stats = meta_clean['file_size'].describe()
print("Image file size stats (bytes):\n", stats, "\n")
if (meta_clean['file_size']==0).any():
    raise ValueError(f"Detected zero-byte image files: {sum(meta_clean['file_size']==0)}")
```

## 5. Feature Engineering & Encoding  
**Purpose & Method:**  
- **Label Encoding** for the target `label` (serialized for later use).  
- **Age Buckets**: categorize `age` into `<30`, `30–45`, `45–60`, `60–90`, `>90` days to capture non-linear effects.  
- **Feature Hashing** (8 dimensions) for `variety` to avoid high-dimensional sparse matrices (serialized).  
- **Standard Scaling** of `age` to standard normal distribution (serialized).  
- Combine all features into a 10 407 × 15 feature matrix.

**Results:**  
- Encoded, scaled, and hashed features ready for modeling.


```python
# Reason: Balancing expressiveness and dimensionality using hashing, bucketing, and scaling.
le = LabelEncoder()
meta_clean['label_enc'] = le.fit_transform(meta_clean['label'])
joblib.dump(le, 'label_encoder.joblib')  # Serialize encoder

# Age buckets
age_max = meta_clean['age'].max(); bins=[0,30,45,60]; labels=['<30','30-45','45-60']
if age_max>90:
    bins += [90, age_max+1]; labels += ['60-90','90+']
else:
    bins += [age_max+1]; labels += ['60+']
meta_clean['age_bucket'] = pd.cut(meta_clean['age'], bins=bins, labels=labels, right=False)
print("Age bucket distribution:\n", meta_clean['age_bucket'].value_counts(), "\n")
age_dummy = pd.get_dummies(meta_clean['age_bucket'], prefix='age')

# Feature hashing for variety
# Transform each variety into a single-element list for the hasher to process correctly
hasher = FeatureHasher(n_features=8, input_type='string')
var_list = meta_clean['variety'].apply(lambda x: [x]).tolist()
var_hash = hasher.transform(var_list)
var_hash_df = pd.DataFrame(var_hash.toarray(), columns=[f'var_hash_{i+1}' for i in range(var_hash.shape[1])])
joblib.dump(hasher, 'feature_hasher.joblib')  # Serialize hasher(hasher, 'feature_hasher.joblib')  # Serialize hasher

# Scaling age
scaler = StandardScaler()
meta_clean['age_scaled'] = scaler.fit_transform(meta_clean[['age']])
joblib.dump(scaler, 'scaler.joblib')  # Serialize scaler

# Combine features
features = pd.concat([
    meta_clean[['image_path','label_enc','age_scaled']],
    age_dummy,
    var_hash_df
], axis=1)
print(f"Feature matrix shape: {features.shape}\n")
```

## 6. Train/Test Split & Resampling  
**Purpose & Method:**  
- Perform a stratified 80/20 train/test split: 8 325 training samples and 2 082 testing samples.  
- Apply SMOTE to the training set within an imblearn pipeline to address class imbalance, expanding to 14 110 samples (pipeline serialized).

**Results:**  
- Balanced training set free from data leakage and an untouched test set for evaluation.


```python
# Reason: Apply resampling only on training set within a pipeline to prevent data leakage.
X = features.drop(['image_path','label_enc'], axis=1)
y = features['label_enc']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
print(f"Train/Test split: {len(y_train)}/{len(y_test)} samples.")

# Pipeline with SMOTE
resample_pipeline = ImbPipeline([('smote', SMOTE(random_state=42))])
X_train_res, y_train_res = resample_pipeline.fit_resample(X_train, y_train)
joblib.dump(resample_pipeline, 'resample_pipeline.joblib')  # Serialize pipeline
print(f"After SMOTE: {len(y_train_res)} training samples (from {len(y_train)}).\n")
```

## 7. Save Processed Data  
**Purpose & Method:**  
- Export processed training and testing sets to CSV.  
- Serialize DataFrames and preprocessing objects (encoders, scalers, pipeline) to ensure full reproducibility.

**Results:**  
- Persistent artifacts facilitating consistent inference and future retraining.

```python
# Reason: Persist processed data and preprocessing objects for future inference/training.
# Save train and test sets
train_df = X_train_res.copy()
train_df['label_enc'] = y_train_res
train_df.to_csv('train_processed_paddy_metadata.csv', index=False)

test_df = X_test.copy()
test_df['label_enc'] = y_test
test_df.to_csv('test_processed_paddy_metadata.csv', index=False)

# Serialize final datasets index for reproducibility
joblib.dump(train_df, 'train_processed_df.joblib')
joblib.dump(test_df, 'test_processed_df.joblib')

print("Preprocessing complete. Objects and datasets serialized for modeling and inference.")
```
